#### How to plot Contour and Surface Plots ####

```python
# Import required modules

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Create a dataset of X and Y values

X = np.array([2,1,4,5,3,6,2,5]).reshape((-1,1))
Y = np.array([1,2,3,5,3,6,4,2]).reshape((-1,1))

print(X.shape,Y.shape)
```

```python
# Let us say the applicable function for this example

func =  X**2 + Y**2
```

```python
# Create a meshgrid - this is a function in numpy

# I have created a bunch of in between data points between 0 and 6 - the max and min values in the data
x = np.arange(0,6,0.5) 
y = np.arange(0,6,0.5)
print(x)

# Translate them in to a meshgrid as below
Xl, Yl = np.meshgrid(x, y)
Xl.shape,Yl.shape
```

```python
# Plot the grid you just created to get a feel for what it looks like 

plt.plot(Xl,Yl, marker='.', color='b', linestyle='none')
plt.show()
```

```python
# Prepare the Z axis value based on the formula X^2 + Y^2 for each data point in the grid
# Reshape it to the shape of the grid

Z =  np.array(Xl**2+Yl**2).reshape(Xl.shape)
print(f"Shape of Z:{Z.shape}\n\nValues of Z{Z}")
```

```python
# Create a surface plot and projected filled contour plot under it.

fig = plt.figure(figsize=(6,5)) # Fix the figure size
ax = fig.gca(projection='3d')   # Generate the 3-D axis 

ax.plot_surface(Xl, Yl, Z, rstride=3,linewidth=1,cmap='viridis')

cset = ax.contourf(Xl, Yl, Z, zdir='z', offset=-0.15, cmap='viridis')

# rotate the axes and update to get different views
angle = 160
a = ax.view_init(15,angle)
```

```python
# The contour plot - plot the data points first and then
# plot the contour based on xl,yl,Z

plt.scatter(X,Y)
CS = plt.contour(Xl,Yl,Z,20)
plt.clabel(CS, inline=1, fontsize=10) # Plots the associated Z values in the plot of 
plt.xlim(0,6)
plt.ylim(0,6)
plt.xlabel("x1")
plt.ylabel("y1")
plt.show()
```
